```python
# from qiskit import QuantumCircuit
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
```

```python
from qiskit import Aer
from qiskit import assemble
```

```python
q = QuantumCircuit(6,6)

# Setup
q.x(0)
q.x(1)
q.x(2)
q.barrier()

# # Fourier
# q.h(0)
# q.h(1)
# q.h(2)
# q.barrier()

# Bramka U

# Pierwszy bit
q.cx(0,3)
q.barrier()

# Åšrodkowy bit
# 1:1

# Ostatni bit
q.cx(1,5)
q.cx(2,5)
q.barrier()

# # Fourier
# q.h(0)
# q.h(1)
# q.h(2)
q.barrier()

# q.measure_all()
q.measure(0,0)
q.measure(1,1)
q.measure(2,2)
q.measure(3,3)
q.measure(4,4)
q.measure(5,5)

q.draw(output='mpl')
```

```python
sym = Aer.get_backend('qasm_simulator')
qobj = assemble(q, shots=1000)
wyniki = sym.run(qobj)

hist = wyniki.result().get_counts()
print(hist)

from qiskit.visualization import plot_histogram

plot_histogram(hist)
```
